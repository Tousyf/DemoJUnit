```python
seasons = ["spring", "summer", "fall", "winter"]
count = 0
for season in seasons:
  print(count, season)
  count +=1
```

```python
season = ["spring", "summer", "fall", "winter"]
for count in range(len(season)):
  print(count,season[count])
```

```python
season1 = ["spring", "summer", "fall", "winter"]
for count,season in enumerate(season1):
  print(count,season)
```

```python
season2 = ["spring", "summer", "fall", "winter"]
for count,season in enumerate(season2, start = 1):
  print(count,season)
```

```python
def fact(n):
  if n == 1:
    return n
  else:
    return n * fact(n-1)
fact(1000)
```

```python
#Greatest common divisor
```

```python
def GCD(x,y):
  rem = x % y
  if rem ==0:
    return y
  else:
    return GCD(y,rem)

n = int(input("enter the first number"))
m = int(input("enter the first number"))
print("The GCD of no. is " , GCD(n,m))
```

```python
#Fibonacci series : Third term should be sum of previous two series
```

```python
def fib(n):
  if n < 2:
    return 1
  else:
    return fib (n-1) + fib(n-2)

n = int(input("enetr the number of terms"))
for i in range(n):
  print(fib(i))
```
