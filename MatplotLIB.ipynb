**Pie Charts**

```python
"""

Univariate/Bi variate Analysis
Categorical vs Numerical
to find contribution on standard scale

we catually draw it on categories and we see how much contribution of category is on 100% scale

"""
```

```python
data = [23, 45, 100, 20, 40]  # contribution of one number on overall sum
plt.pie(data)
plt.show()
```

```python
data = [23, 45, 100, 20, 40]
subjects = ["science", "hindi", "english", "sst", "maths"]
plt.pie(data, labels = subjects)

plt.show()
```
